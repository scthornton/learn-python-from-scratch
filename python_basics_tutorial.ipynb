## 7. Conditionals: Making Decisions

`if` statements let your code make decisions based on conditions.

```python
# Basic if statement
temperature = 75

if temperature > 80:
    print("It's hot outside!")
elif temperature > 60:
    print("The weather is nice.")
else:
    print("It's cold outside!")
```

```python
# Comparison operators
x = 10
y = 5

print("x == y:", x == y)  # Equal to
print("x != y:", x != y)  # Not equal to
print("x > y:", x > y)    # Greater than
print("x < y:", x < y)    # Less than
print("x >= y:", x >= y)  # Greater than or equal to
print("x <= y:", x <= y)  # Less than or equal to
```

```python
# Practical example: checking age for movie rating
age = 16
movie_rating = "R"

if movie_rating == "G":
    print("Everyone can watch!")
elif movie_rating == "PG-13" and age >= 13:
    print("You can watch this movie.")
elif movie_rating == "R" and age >= 17:
    print("You can watch this movie.")
else:
    print("You're too young for this movie.")
```

## 8. Loops: Repeating Actions

Loops let you repeat code multiple times without writing it over and over.

```python
# For loop: repeat for each item in a list
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print("I like", fruit)
```

```python
# For loop with range: repeat a specific number of times
for i in range(5):
    print("Count:", i)  # Prints 0, 1, 2, 3, 4
```

```python
# While loop: repeat as long as a condition is true
count = 0

while count < 5:
    print("Count is:", count)
    count = count + 1  # Increase count by 1
```

```python
# Practical example: calculating total price
prices = [12.99, 8.50, 15.75, 22.00]
total = 0

for price in prices:
    total = total + price

print("Total price:", total)
```

## 9. Functions: Reusable Code Blocks

Functions are named blocks of code that you can use over and over.

```python
# Define a function using 'def'
def greet(name):
    print("Hello,", name + "!")

# Call the function
greet("Alice")
greet("Bob")
greet("Charlie")
```

```python
# Functions can return values
def add_numbers(a, b):
    result = a + b
    return result

sum1 = add_numbers(5, 3)
sum2 = add_numbers(10, 20)

print("Sum 1:", sum1)
print("Sum 2:", sum2)
```

```python
# Practical example: calculate tip function
def calculate_tip(bill, tip_percentage):
    tip = bill * tip_percentage
    total = bill + tip
    return total

# Use the function for different bills
dinner = calculate_tip(45.50, 0.20)
lunch = calculate_tip(12.00, 0.15)

print("Dinner total:", dinner)
print("Lunch total:", lunch)
```

## 10. Working with Strings

Strings (text) have many useful methods for manipulation.

```python
# String basics
text = "Hello, Python!"

print("Original:", text)
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
print("Length:", len(text))
```

```python
# String methods
sentence = "  Python is awesome!  "

print("Original:", sentence)
print("Stripped:", sentence.strip())  # Remove extra spaces
print("Replace:", sentence.replace("awesome", "amazing"))
print("Starts with 'Python':", sentence.strip().startswith("Python"))
print("Contains 'is':", "is" in sentence)
```

```python
# String formatting (f-strings)
name = "Alice"
age = 25
height = 5.7

# Modern way to combine variables and text
message = f"My name is {name}, I am {age} years old, and {height} feet tall."
print(message)

# You can even do calculations inside f-strings
years_until_30 = f"I will be 30 in {30 - age} years."
print(years_until_30)
```

```python
# Splitting and joining strings
csv_data = "apple,banana,orange,grape"
fruits = csv_data.split(",")  # Split into list
print("List:", fruits)

joined = " | ".join(fruits)  # Join list into string
print("Joined:", joined)
```

## 12. Working with JSON (Common Data Format)

JSON is a popular format for storing and exchanging data. Python makes it easy to work with.

```python
import json

# Python dictionary
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York",
    "hobbies": ["reading", "hiking", "coding"]
}

# Convert to JSON string
json_string = json.dumps(person, indent=2)
print("JSON format:")
print(json_string)
```

```python
# Convert JSON string back to Python dictionary
json_data = '{"name": "Bob", "age": 30, "city": "Boston"}'
person_dict = json.loads(json_data)

print("Python dictionary:")
print(person_dict)
print("Name:", person_dict["name"])
print("Age:", person_dict["age"])
```

```python
# Save dictionary to JSON file
data = {
    "users": [
        {"name": "Alice", "age": 25},
        {"name": "Bob", "age": 30},
        {"name": "Charlie", "age": 35}
    ]
}

with open("users.json", "w") as file:
    json.dump(data, file, indent=2)

print("JSON file saved!")
```

```python
# Read dictionary from JSON file
with open("users.json", "r") as file:
    loaded_data = json.load(file)

print("Loaded data:")
for user in loaded_data["users"]:
    print(f"{user['name']} is {user['age']} years old")
```

## 13. Error Handling: Dealing with Mistakes

Sometimes code encounters errors. You can handle them gracefully using try/except.

```python
# Without error handling - this would crash
# number = int("not a number")  # This line would cause an error
```

```python
# With error handling
try:
    number = int("not a number")
    print("Number:", number)
except ValueError:
    print("Error: That's not a valid number!")
```

```python
# Practical example: safe division
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero!"

print(safe_divide(10, 2))   # Works fine
print(safe_divide(10, 0))   # Handles error gracefully
```

## 14. List Comprehensions: Elegant List Creation

List comprehensions are a concise way to create lists.

```python
# Traditional way
squares = []
for i in range(10):
    squares.append(i * i)

print("Traditional:", squares)
```

```python
# List comprehension way (same result, cleaner code)
squares = [i * i for i in range(10)]
print("List comprehension:", squares)
```

```python
# With conditions
even_numbers = [i for i in range(20) if i % 2 == 0]
print("Even numbers:", even_numbers)

# Transforming strings
names = ["alice", "bob", "charlie"]
uppercase_names = [name.upper() for name in names]
print("Uppercase names:", uppercase_names)
```

## 15. Working with APIs: Making HTTP Requests

APIs let programs talk to each other over the internet. The `requests` library makes this easy.

```python
import requests

# Make a GET request to a public API
response = requests.get("https://api.github.com/users/github")

print("Status code:", response.status_code)  # 200 means success
print("Response type:", type(response.json()))
```

## 16. Putting It All Together: A Practical Example

Let's build a simple program that manages a todo list.

```python
import json
from datetime import datetime

class TodoList:
    def __init__(self):
        self.tasks = []
    
    def add_task(self, task):
        """Add a new task to the list"""
        self.tasks.append({
            "task": task,
            "completed": False,
            "created": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })
        print(f"Added: {task}")
    
    def complete_task(self, task_number):
        """Mark a task as completed"""
        if 0 <= task_number < len(self.tasks):
            self.tasks[task_number]["completed"] = True
            print(f"Completed: {self.tasks[task_number]['task']}")
        else:
            print("Invalid task number")
    
    def show_tasks(self):
        """Display all tasks"""
        if not self.tasks:
            print("No tasks yet!")
            return
        
        print("\n=== TODO LIST ===")
        for i, task in enumerate(self.tasks):
            status = "âœ“" if task["completed"] else " "
            print(f"{i}. [{status}] {task['task']}")
    
    def save_to_file(self, filename):
        """Save tasks to a JSON file"""
        with open(filename, "w") as file:
            json.dump(self.tasks, file, indent=2)
        print(f"Saved to {filename}")
    
    def load_from_file(self, filename):
        """Load tasks from a JSON file"""
        try:
            with open(filename, "r") as file:
                self.tasks = json.load(file)
            print(f"Loaded {len(self.tasks)} tasks from {filename}")
        except FileNotFoundError:
            print(f"File {filename} not found")
```

```python
# Use the TodoList
todo = TodoList()

# Add some tasks
todo.add_task("Learn Python basics")
todo.add_task("Build a project")
todo.add_task("Share my work")

# Show tasks
todo.show_tasks()
```

```python
# Complete a task
todo.complete_task(0)
todo.show_tasks()
```

```python
# Save and load
todo.save_to_file("my_todos.json")

# Create a new list and load the saved data
new_todo = TodoList()
new_todo.load_from_file("my_todos.json")
new_todo.show_tasks()
```
